```python
import matplotlib.pyplot as plt

# Data dummy (hanya sebagian untuk contoh)
wilayah = ["Baktiya", "Baktiya Barat", "Banda Baro", "Cot Girek", "Dewantara", "Muara Batu"]
fatalities = [6, 9, 3, 7, 5, 9]
accidents = [8, 13, 4, 11, 6, 14]
clusters = [1, 3, 2, 3, 1, 3]  # Hasil klaster
medoids = ["Baktiya", "Banda Baro", "Muara Batu"]
medoid_fatalities = [6, 3, 9]
medoid_accidents = [8, 4, 14]

plt.figure(figsize=(10, 6))
colors = ['blue', 'green', 'red']
for i in range(1, 4):
    cluster_fatalities = [f for f, c in zip(fatalities, clusters) if c == i]
    cluster_accidents = [a for a, c in zip(accidents, clusters) if c == i]
    plt.scatter(cluster_fatalities, cluster_accidents, c=colors[i-1], label=f'Klaster {i}')

# Tandai medoid
plt.scatter(medoid_fatalities, medoid_accidents, c='black', marker='*', s=200, label='Medoid')

# Label dan judul
plt.title('Klusterisasi Wilayah Aceh Utara dengan K-Medoids')
plt.xlabel('Fatalities')
plt.ylabel('Accidents')
plt.legend()
plt.grid(True)

# Simpan gambar
plt.savefig('kmedoids_clustering.png')
```
